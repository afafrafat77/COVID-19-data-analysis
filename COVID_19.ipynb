# Merge the two dataframes


```python
merged_df = country_wise_df.merge(covid_complete_df, on='Country', how='inner')
merged_df
```

```python
# Calculate average values for Confirmed, Deaths, Recovered, and Active cases
merged_df['Confirmed'] = (merged_df['Confirmed_x'] + merged_df['Confirmed_y']) / 2
merged_df['Deaths'] = (merged_df['Deaths_x'] + merged_df['Deaths_y']) / 2
merged_df['Recovered'] = (merged_df['Recovered_x'] + merged_df['Recovered_y']) / 2
merged_df['Active'] = (merged_df['Active_x'] + merged_df['Active_y']) / 2

# Drop redundant columns
merged_df.drop(columns=['Confirmed_x', 'Deaths_x', 'Recovered_x', 'Active_x', 'Confirmed_y', 'Deaths_y', 'Recovered_y', 'Active_y'], inplace=True)
merged_df.drop_duplicates(inplace=True)
merged_df.drop(columns=['WHO Region_y'], inplace=True)
merged_df.rename(columns={'WHO Region_x': 'WHO Region'}, inplace=True)
merged_df
```

```python
merged_df.dtypes
```

```python
merged_df['WHO Region'] = merged_df['WHO Region'].astype('category')
merged_df
```

```python
# Retain rows where 'Confirmed' is between 10 and 1 million
merged_df = merged_df[(merged_df['Confirmed'] >= 10) & (merged_df['Confirmed'] < 1000000)]
```

```python
merged_df.drop_duplicates(inplace=True)
merged_df
```

```python
country_data =merged_df .groupby('Country')[['Confirmed', 'Deaths', 'Recovered']].sum()
country_data
```

```python
daily_data = merged_df .groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum()
daily_data
```
